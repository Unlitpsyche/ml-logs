```python
#1
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace()
y = 2*x + 3
plt.plot(x, y)
plt.title('y = 2x + 3')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

```python
#2
x = np.linspace(-10, 10, 100)
y = x**2
plt.plot(x, y, 'r--', linewidth=2, marker='o')
plt.show()
```

```python
#3
x = np.linspace(-10, 10, 100)
plt.plot(x, x**2, 'g-', label='x^2')
plt.plot(x, x**3, 'b--', label='x^3')
plt.legend()
plt.show()
```

```python
#4
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

fig, axs = plt.subplots(1, 2, figsize=(10, 4))

axs[0].plot(x, y_sin)
axs[0].set_title('sin(x)')

axs[1].plot(x, y_cos)
axs[1].set_title('cos(x)')

plt.tight_layout()
plt.show()
```

```python
#6
products = ['A', 'B', 'C', 'D']
sales = [120, 150, 90, 200]
plt.bar(products, sales)
plt.xlabel('Products')
plt.ylabel('Sales')
plt.show()
```

```python
#7
idx = np.argsort(sales)[::-1]
plt.barh(np.array(products)[idx], np.array(sales)[idx])
plt.xlabel('Sales')
plt.ylabel('Products')
plt.show()
```

```python
#8
np.random.seed(1)
data = np.random.normal(50, 5, 1000)
plt.hist(data, bins=20)
plt.grid()
plt.show()
```

```python
#9
import matplotlib.pyplot as plt

labels = ['Python', 'Java', 'C++', 'JavaScript']
usage = [40, 30, 15, 15]
explode = [0.1 if i == max(usage) else 0 for i in usage]

plt.pie(usage, labels=labels, explode=explode, autopct='%1.1f%%', shadow=True)
plt.title('Programming Language Usage')
plt.show()
```

```python
#10
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv(r"C:\Users\Dell\dms_68d0b7882d78fcompany_sales_data.csv")
plt.plot(data['month_number'], data['total_profit'])

plt.xlabel('Month Number')
plt.ylabel('Total profit')
plt.title('Company Total Profit per Month')
plt.show()
```

```python
#11
plt.plot(data['month_number'], data['total_profit'],
         linestyle=':', color='red', marker='o', linewidth=3, label='Total Profit')

plt.xlabel('Month Number')
plt.ylabel('Total profit')
plt.title('Company Total Profit per Month')
plt.legend(loc='upper left')
plt.show()
```

```python
#12
products = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']

for product in products:
    plt.plot(data['month_number'], data[product], label=product)

plt.xlabel('Month Number')
plt.ylabel('Sold units number')
plt.title('Units Sold Per Month for Each Product')
plt.legend()
plt.show()
```

```python
#13
import numpy as np

bar_width = 0.35
index = np.arange(len(data['month_number']))

plt.bar(index, data['facecream'], bar_width, label='Face Cream')
plt.bar(index + bar_width, data['facewash'], bar_width, label='Face Wash')

plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.title('Sales of Face Cream and Face Wash per Month')
plt.xticks(index + bar_width / 2, data['month_number'])
plt.legend()
plt.show()
```

```python
#14
plt.bar(data['month_number'], data['bathingsoap'])
plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.title('Bathing Soap Sales per Month')
plt.savefig('bathing_soap_sales.png')
plt.show()
plt.close()
```

```python
#15
import seaborn as sns
import matplotlib.pyplot as plt

products_data = data[['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']]
corr = products_data.corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Product Sales Correlation Heatmap')
plt.show()
```
